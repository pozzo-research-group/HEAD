```python
import numpy as np
import matplotlib.pyplot as plt
from SAXS_model import model
```

```python
q = np.linspace(1e-1, 1,1000)
Sphere_rad = 6 #nm
Sphere_std = 0.5 #nm
I = model(q,Sphere_rad, Sphere_std)
```

```python
fig, ax = plt.subplots()
ax.plot(q, I)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('$ q \ (nm^{-1})$')
ax.set_ylabel('Intensity (a.u)')
```
